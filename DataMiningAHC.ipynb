```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, single, complete, average, linkage
```

```python
data = pd.read_excel("Data AHC.xlsx")
print(data)
```

```python
x = np.asarray(data)
print(x)
```

```python
x = np.asarray(data)
print(x)
```

```python
x = average(x)
y = single(x)
z = complete(x)
print(z)
print(y)
print(x)
```

```python
plt.figure(figsize=(25,10))
plt.title('Dendrogram Hierarchical Klastering, complete Linkage')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(z)
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.title('Dendrogram Hierarchical Klastering, single linkage')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(y)
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.title('Dendrogram Hierarchical Klastering, average linkage')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(x)
plt.show()
```
